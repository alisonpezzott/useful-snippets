```python
import sempy_labs as labs
from sempy_labs.report import ReportWrapper
import sempy.fabric as fabric
```

```python
ws = '0d4a2f53-8c7b-42b0-8500-da59d8da80df'
ds = 'cbe64d79-1449-4d67-96ea-1e858a5253cf'
```

```python
# Objects to refresh, define using a dictionary
objects_to_refresh = [
    {
        "table": "Vendas",
        "partition": "2019"
    }
]

# Refresh the dataset
fabric.refresh_dataset(
    workspace=ws,
    dataset=ds, 
    objects=objects_to_refresh,
    apply_refresh_policy=False
)
```

```python
# List the refresh requests
df = fabric.list_refresh_requests(dataset=ds, workspace=ws)
df_filtrado = df[df['Request Id'] == '9ce4ce7e-f799-4357-b664-459663ec1949']
display(df_filtrado)

print(df.loc[0, 'Service Exception Json'])
print(df.loc[0, 'Refresh Attempts'])
```
