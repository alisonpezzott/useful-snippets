```python
%%tsql -artifact DB_Sandbox -type SQLDatabase
create table cidades(cidade varchar(200), uf varchar(2));
```

```python
dados = [
    ('Jundiaí', 'SP'),
    ('Piracicaba', 'SP'),
    ('Várzea Paulista', 'SP')
]

query = """
insert into cidades (cidade, uf) 
values ('{cidade}', '{uf}');
"""

for cidade, uf in dados:
    %tsql -artifact DB_Sandbox -type SQLDatabase {query} 


%tsql -artifact DB_Sandbox -type SQLDatabase select cidade, uf from cidades
```
